```python
img = cv2.imread('./robot_man.jpg') # reads image in BGR
img.shape
```

```python
imshow(img)
```

```python
h,l = img.shape[0],img.shape[1]
```

```python
color_dict = {}
color_keys = []
for i in range(h):
    for j in range(l):
        [B,G,R] = img[i][j]
        color_key = [ (B//10)*10,(G//10)*10,(R//10)*10 ]
        color_keys.append(color_key)
        color_dict[tuple(color_key)] = color_dict.get(tuple(color_key) ,0) + 1
        
        
```

```python
color_keys = [list(t) for t in color_dict.keys()]
```

```python
K = 7  # Replace with your desired number of clusters
kmeans = KMeans(n_clusters=K)
cluster_assignments = kmeans.fit_predict(color_keys)
```

```python
img1 = cv2.imread('./robot_man.jpg')
count1= 0
for i in range(h):
    for j in range(l):
        [B,G,R] = img1[i][j]
        color_key = [ [(B//10)*10,(G//10)*10,(R//10)*10] ]
        if kmeans.predict(color_key)==1:
            count1 +=1
            img1[i][j] = [256 , 256, 256]

        

plt.imshow(img1)
plt.axis('off')  
plt.show()
        
```
